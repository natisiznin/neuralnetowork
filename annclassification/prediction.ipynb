```python
import tensorflow as tf
import tensorflow.keras.models as models
import pickle
import pandas as pd 
import numpy as np
```

```python
model = models.load_model('model.h5')

with open('label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)
with open('oh_encoder.pkl', 'rb') as f:
    oh_encoder = pickle.load(f)
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
```

```python
input_data={
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 5000

}
```

```python
geo_encoded = oh_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=oh_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df['Gender'] = label_encoder.transform(input_df['Gender'])
```

```python
input_df = pd.concat([input_df.drop(columns=['Geography']), geo_encoded_df], axis=1)
```

```python
input_df
```

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_prob = prediction[0][0]
prediction_prob
```

```python
if prediction_prob > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is unlikely to churn.")
```
